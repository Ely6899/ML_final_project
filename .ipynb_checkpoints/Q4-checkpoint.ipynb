```python
!pip -q install torch torchvision scikit-learn tqdm seaborn matplotlib

import os, torch, numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
from pathlib import Path
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, fbeta_score, confusion_matrix)
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.preprocessing import StandardScaler
from tqdm.auto import tqdm

# ---------- pick safest device ----------
def safe_device():
    if not torch.cuda.is_available(): return "cpu"
    try:
        torch.zeros(1).to("cuda"); return "cuda"
    except RuntimeError:
        return "cpu"
DEVICE = safe_device()
print("🖥  Using device:", DEVICE.upper())
```

```python
root   = Path("/content/country15")  # change if needed
TARGET = ['US','IL','JP','NG','DE']

# original → new idx
full = datasets.ImageFolder(root/'train')
cls2idx = {c:i for i,c in enumerate(full.classes)}
keep    = [cls2idx[c] for c in TARGET]
lut     = {old:new for new,old in enumerate(keep)}

tf_train = transforms.Compose([
    transforms.Resize(224), transforms.RandomHorizontalFlip(),
    transforms.CenterCrop(224), transforms.ToTensor(),
    transforms.Normalize([.485,.456,.406],[.229,.224,.225])
])
tf_eval = transforms.Compose([
    transforms.Resize(224), transforms.CenterCrop(224), transforms.ToTensor(),
    transforms.Normalize([.485,.456,.406],[.229,.224,.225])
])

def subset(folder, tf):
    ds = datasets.ImageFolder(folder, tf)
    ds.samples = [(p, lut[l]) for p,l in ds.samples if l in keep]
    return ds

train_ds = subset(root/'train', tf_train)
val_ds   = subset(root/'valid', tf_eval)
test_ds  = subset(root/'test' , tf_eval)

dl = dict(num_workers=2, pin_memory=(DEVICE=="cuda"))
train_ld = DataLoader(train_ds, 32, shuffle=True , **dl)
val_ld   = DataLoader(val_ds  , 64, shuffle=False, **dl)
test_ld  = DataLoader(test_ds , 64, shuffle=False, **dl)
print("Samples →", len(train_ds), len(val_ds), len(test_ds))
```

```python
res18 = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1).eval()
res_feat = torch.nn.Sequential(*(list(res18.children())[:-1]), torch.nn.Flatten()).to(DEVICE)

def res_feats(loader):
    Z, y = [], []
    with torch.no_grad():
        for xb, yb in tqdm(loader, desc="ResNet-feat"):
            Z.append(res_feat(xb.to(DEVICE)).cpu()); y.append(yb)
    return torch.cat(Z).numpy(), torch.cat(y).numpy()

Xtr_r, ytr = res_feats(train_ld)
Xva_r, yva = res_feats(val_ld)
Xte_r, yte = res_feats(test_ld)
sc_r = StandardScaler().fit(Xtr_r)
Xtr_r, Xva_r, Xte_r = sc_r.transform(Xtr_r), sc_r.transform(Xva_r), sc_r.transform(Xte_r)
```

```python
cache = Path("/content/vgg_emb.npz")
if cache.exists():
    data = np.load(cache)
    Xtr_v, ytr_v = data['Xtr'], data['ytr']
    Xva_v, yva_v = data['Xva'], data['yva']
    Xte_v, yte_v = data['Xte'], data['yte']
    print("✅ VGG cache loaded:", Xtr_v.shape)
else:
    vgg = models.vgg16(weights=models.VGG16_Weights.IMAGENET1K_V1).to(DEVICE).eval()
    vgg_feat = torch.nn.Sequential(*vgg.features,
                                   torch.nn.AdaptiveAvgPool2d((7,7)),
                                   torch.nn.Flatten(),
                                   *list(vgg.classifier)[:-1])
    def embed(loader):
        Z, y = [], []
        with torch.no_grad():
            for xb, yb in tqdm(loader, desc="VGG-emb"):
                Z.append(vgg_feat(xb.to(DEVICE)).cpu()); y.append(yb)
        return torch.cat(Z).numpy(), torch.cat(y).numpy()
    Xtr_v, ytr_v = embed(train_ld); Xva_v, yva_v = embed(val_ld); Xte_v, yte_v = embed(test_ld)
    np.savez_compressed(cache, Xtr=Xtr_v, ytr=ytr_v, Xva=Xva_v, yva=yva_v, Xte=Xte_v, yte=yte_v)

sc_v = StandardScaler().fit(Xtr_v)
Xtr_v, Xva_v, Xte_v = sc_v.transform(Xtr_v), sc_v.transform(Xva_v), sc_v.transform(Xte_v)
```

```python
# --- 5. Colour histograms (fixed) -----------------------------------------
from PIL import Image
from tqdm.auto import tqdm
import numpy as np
from sklearn.preprocessing import StandardScaler

def rgb_hist(dataset, bins=256, size=96):
    """
    Returns an (N, bins*3) array of RGB histograms.
    Works on CPU, no torch ops inside.
    """
    feats = []
    for path, _ in tqdm(dataset.samples, desc="RGB-hist"):
        img  = Image.open(path).convert("RGB").resize((size, size))
        arr  = np.asarray(img)                  # (H,W,3)
        hist = np.concatenate([
            np.histogram(arr[:, :, c], bins=bins, range=(0, 255))[0]
            for c in range(3)
        ])
        feats.append(hist.astype(np.float32))
    return np.stack(feats)

Xtr_rgb = rgb_hist(train_ds)
Xva_rgb = rgb_hist(val_ds)
Xte_rgb = rgb_hist(test_ds)

# standardise
sc_h = StandardScaler().fit(Xtr_rgb)
Xtr_rgb = sc_h.transform(Xtr_rgb)
Xva_rgb = sc_h.transform(Xva_rgb)
Xte_rgb = sc_h.transform(Xte_rgb)

print("RGB-hist shapes:", Xtr_rgb.shape, Xva_rgb.shape, Xte_rgb.shape)
```

```python
def log_metrics(name, model, Xtr, ytr, Xte, yte, store):
    model.fit(Xtr, ytr); yp = model.predict(Xte)
    store.append(dict(
        model=name,
        acc = accuracy_score(yte, yp),
        prec= precision_score(yte, yp, average='macro', zero_division=0),
        rec = recall_score(yte, yp, average='macro', zero_division=0),
        f1  = f1_score(yte, yp, average='macro', zero_division=0),
        f2  = fbeta_score(yte, yp, beta=2, average='macro', zero_division=0)
    ))
    cm = confusion_matrix(yte, yp, normalize='true')
    sns.heatmap(cm, annot=True, fmt=".2f",
                xticklabels=TARGET, yticklabels=TARGET, cmap="rocket_r")
    plt.title(name); plt.show()
```

```python
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, fbeta_score
import pandas as pd

results = []

def log_metrics(name, model, X_train, y_train, X_test, y_test, result_list):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc  = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average='macro')
    rec  = recall_score(y_test, y_pred, average='macro')
    f1   = f1_score(y_test, y_pred, average='macro')
    f2   = fbeta_score(y_test, y_pred, average='macro', beta=2)
    result_list.append(dict(model=name, acc=acc, prec=prec, rec=rec, f1=f1, f2=f2))

# Dataset 1: ResNet and RGB histogram
for tag, Xtr, Xte in [("ResNet", Xtr_r, Xte_r),
                      ("RGB",    Xtr_rgb, Xte_rgb)]:
    log_metrics(f"SVM-{tag}", LinearSVC(max_iter=8000), Xtr, ytr, Xte, yte, results)
    log_metrics(f"RF-{tag}",  RandomForestClassifier(n_estimators=400), Xtr, ytr, Xte, yte, results)
    log_metrics(f"ADA-{tag}", AdaBoostClassifier(n_estimators=400), Xtr, ytr, Xte, yte, results)

# Dataset 2: VGG embeddings
for tag, Xtr, Xte in [("VGG", Xtr_v, Xte_v)]:
    log_metrics(f"SVM-{tag}", LinearSVC(max_iter=8000), Xtr, ytr_v, Xte, yte_v, results)
    log_metrics(f"RF-{tag}",  RandomForestClassifier(n_estimators=400), Xtr, ytr_v, Xte, yte_v, results)
    log_metrics(f"ADA-{tag}", AdaBoostClassifier(n_estimators=400), Xtr, ytr_v, Xte, yte_v, results)

# Show as table
df = pd.DataFrame(results).sort_values('acc', ascending=False)
df.style.background_gradient(cmap='Blues', subset=['acc', 'f1', 'f2'])


```

```python
df = pd.DataFrame(results).set_index('model')
display(df.style.format('{:.2%}').set_caption("Macro metrics"))

# optional radar chart visualization
def radar(df_row):
    labels = ['acc','prec','rec','f1','f2']
    vals = df_row[labels].values
    theta = np.linspace(0, 2 * np.pi, len(labels), endpoint=False)

    # wrap-around for closed radar
    vals_wrapped = np.concatenate((vals, [vals[0]]))
    theta_wrapped = np.concatenate((theta, [theta[0]]))

    plt.figure(figsize=(4, 4))
    plt.polar(theta_wrapped, vals_wrapped, marker='o')
    plt.fill(theta_wrapped, vals_wrapped, alpha=0.25)
    plt.thetagrids(theta * 180 / np.pi, labels)  # <- keep original 5 angles for labels
    plt.ylim(0, 1)
    plt.title(df_row.name)
    plt.show()

df.apply(radar, axis=1)
```

### 📌 Key Findings

| 🔍 Insight | 📈 Observation |
|-----------|----------------|
| 🎨 **RGB histograms underperform** | All models on RGB features gave poor results (~17–20% accuracy), showing that color alone isn’t enough for distinguishing countries. |
| 🧠 **Deep features are stronger** | ResNet/VGG embeddings improved model accuracy to ~33–36%, confirming the importance of semantic texture/scene information. |
| 🔁 **VGG slightly edges out ResNet** | On average, VGG embeddings gave the best macro scores (accuracy, F1/F2). |
| 🔧 **Model performance is stable** | Random Forest and SVM consistently outperformed AdaBoost, but differences were modest. |

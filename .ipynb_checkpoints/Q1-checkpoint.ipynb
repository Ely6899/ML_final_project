# üåç Country-15 Similarity Study
We will answer two questions  

1. **Which countries‚Äô images look the most alike?**  
2. **Which are the most different?**

Strategy

| Stage | Tool / Model | Purpose |
|-------|--------------|---------|
| Pre-processing | `torchvision.transforms` | Standard resize / normalise |
| Feature extraction | **ResNet-50**, **VGG-16** (pre-trained, frozen) | Rich, general visual embeddings |
| Dimensionality cut | Optional **PCA** | Scatter-plots & faster classifiers |
| Classifiers | Linear **SVM**, **RandomForest**, **AdaBoost** | Produce confusion matrices |
| Similarity metrics | ‚Ä¢ Euclidean distance between class centroids in embedding space  <br>‚Ä¢ ‚ÄúMix-up score‚Äù = off-diagonal confusion-rate | Rank most-similar / most-different pairs |

```python
def pick_device():
    if not torch.cuda.is_available():
        return "cpu"
    try:
        # 1-byte tensor test ‚Üí catches bad GPU states
        torch.zeros(1).to("cuda")
        return "cuda"
    except RuntimeError as e:
        print("‚ö†Ô∏è  GPU not usable, switching to CPU:", e)
        return "cpu"

DEVICE = pick_device()
print("Using device:", DEVICE)
```

```python
from pathlib import Path
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

root   = Path("/content/country15")      # ‚Üê adjust if needed
TARGET = ['IL','US','JP','NG','DE']      # five classes

raw_tf   = transforms.ToTensor()
full_tr  = datasets.ImageFolder(root/'train', raw_tf)
full_val = datasets.ImageFolder(root/'valid', raw_tf)
full_te  = datasets.ImageFolder(root/'test' , raw_tf)

cls2idx  = {c:i for i,c in enumerate(full_tr.classes)}
keep_idx = [cls2idx[c] for c in TARGET]
lut      = {old:new for new,old in enumerate(keep_idx)}  # map old‚Üí0-4

def filter_remap(ds):
    ds.samples = [(p, lut[l]) for p,l in ds.samples if l in keep_idx]
    return ds

train_ds = filter_remap(full_tr)
val_ds   = filter_remap(full_val)
test_ds  = filter_remap(full_te)

print("Counts  ‚Üí", len(train_ds), len(val_ds), len(test_ds))
print("Label set‚Üí", sorted({l for _,l in test_ds.samples}))
```

```python
from torchvision import transforms
IMG = 96
tx_train = transforms.Compose([
    transforms.Resize((IMG,IMG)), transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10), transforms.ToTensor(),
    transforms.Normalize([.485,.456,.406],[.229,.224,.225])
])
tx_eval  = transforms.Compose([
    transforms.Resize((IMG,IMG)), transforms.ToTensor(),
    transforms.Normalize([.485,.456,.406],[.229,.224,.225])
])

train_ds.transform = tx_train
val_ds.transform   = test_ds.transform = tx_eval

train_loader = DataLoader(train_ds,32,shuffle=True ,num_workers=2)
test_loader  = DataLoader(test_ds ,64,shuffle=False,num_workers=2)
```

```python
import torch.nn as nn, numpy as np, tqdm, torch

@torch.no_grad()
def extract_feats(model, loader, device):
    model = model.eval().to(device)
    feats, labels = [], []
    for xb,yb in tqdm.tqdm(loader, desc=f"Extract@{device}"):
        xb = xb.to(device, non_blocking=True)
        feats.append(model(xb).cpu())
        labels.append(yb)
    return torch.cat(feats).numpy(), torch.cat(labels).numpy()
```

```python
from torchvision import models

resnet = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V2)
resnet_back = nn.Sequential(*(list(resnet.children())[:-1]), nn.Flatten())

vgg = models.vgg16(weights=models.VGG16_Weights.IMAGENET1K_V1)
vgg_back = nn.Sequential(*vgg.features, nn.AdaptiveAvgPool2d(1), nn.Flatten())
```

```python
resnet_feats, y_test = extract_feats(resnet_back, test_loader, DEVICE)
vgg_feats,   _       = extract_feats(vgg_back , test_loader, DEVICE)
print("Shapes:", resnet_feats.shape, vgg_feats.shape)
```

```python
from scipy.spatial.distance import pdist, squareform
import numpy as np, seaborn as sns, matplotlib.pyplot as plt

def centroid_matrix(feats, labels):
    C = np.vstack([feats[labels == i].mean(0) for i in range(len(TARGET))])
    return squareform(pdist(C, 'euclidean')), C

D, centroids = centroid_matrix(resnet_feats, y_test)

# Compute off-diagonal mask (excluding same-class comparisons)
mask = np.triu(np.ones_like(D), k=1).astype(bool)
off_diag = D[mask]

# Get index of most similar & most different off-diagonal pairs
min_idx = np.argmin(off_diag)
max_idx = np.argmax(off_diag)

# Map back to 2D index from upper triangle
pairs = [(i, j) for i in range(len(TARGET)) for j in range(i+1, len(TARGET))]
most_sim = pairs[min_idx]
most_diff = pairs[max_idx]

print(f"‚úÖ Most similar countries   : {TARGET[most_sim[0]]} & {TARGET[most_sim[1]]} (distance = {D[most_sim]:.2f})")
print(f"‚úÖ Most different countries : {TARGET[most_diff[0]]} & {TARGET[most_diff[1]]} (distance = {D[most_diff]:.2f})")

# Heatmap visualization
sns.heatmap(D, annot=True, fmt=".2f", xticklabels=TARGET, yticklabels=TARGET, cmap="Blues")
plt.title("ResNet centroid distances (corrected)")
plt.show()
```

```python
import networkx as nx

# Build graph from pairwise distances
G = nx.Graph()
for i in range(len(TARGET)):
    G.add_node(TARGET[i])

# Add edges with inverse-distance weights (closer = thicker)
for (i, j) in pairs:
    dist = D[i, j]
    G.add_edge(TARGET[i], TARGET[j], weight=1.0 / (dist + 1e-5))

# Plot it
pos = nx.spring_layout(G, seed=42)  # fixed layout
plt.figure(figsize=(7,6))
edges, weights = zip(*nx.get_edge_attributes(G, 'weight').items())

nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=1000,
        width=[w * 8 for w in weights], edge_color='gray', font_size=14)

plt.title("üåç Visual similarity graph (closer = thicker)")
plt.show()
```
